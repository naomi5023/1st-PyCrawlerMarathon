### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
/cwbopendata/dataset/locations/location[19]/weatherElement[5]/description
'''
import xmltodict

# 存取檔案

with open('./example/64_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊
root = doc['cwbopendata']
# /cwbopendata/dataset/locations/location[2]/locationName
print('高雄市總共有幾個地區資料: ', end='')
print(len(root['dataset']['locations']['location']))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
locations = root['dataset']['locations']['location']
for location in locations:
    print(location['locationName'])
    for weatherElement in location['weatherElement']:
        if weatherElement['description'] == '溫度':
            print(weatherElement['time'][0]['dataTime'])
            print(weatherElement['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
first_location = locations[0]
print(first_location['locationName'])
for weatherElement in first_location['weatherElement']:
    if weatherElement['description'] == '溫度':
        time_points = weatherElement['time'] 
        for time in time_points:
            print(time['dataTime'])
            print(time['elementValue']['value'])
            print()
        
    
```
