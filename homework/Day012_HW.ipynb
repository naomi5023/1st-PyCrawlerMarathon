## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup
```

### ① 取出今天所有的發文

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
articles = dict()
for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find(class_="tag").text
    title = d.find_all('a')[-1].text
    if category not in articles:
        articles[category] = list()
    articles[category].append(title)
```

### ③ 哪一個類別的文章最多

```python
count = {(category, len(titles)) for (category,titles) in articles.items()}
print(count)
# 印出資料後發現 影劇最多
```
